# 1 模型初始化

## 1.1 使用ChatOpenAI初始化

```python
import os
from langchain_openai import ChatOpenAI

api_key = os.getenv("HUNYUAN_API_KEY")

model = ChatOpenAI(
    model="hunyuan-lite",
    base_url="https://api.hunyuan.cloud.tencent.com/v1",
    api_key=api_key
)

response = model.invoke("你是哪个模型")
print(response.content)
```

## 1.2 使用init_chat_model初始化

```python
import os
from langchain.chat_models import init_chat_model

api_key = os.getenv("HUNYUAN_API_KEY")

model = init_chat_model(
    model="hunyuan-lite",
    model_provider="openai",
    api_key=api_key,
    base_url="https://api.hunyuan.cloud.tencent.com/v1",
    temperature=1.5,
)

response = model.invoke("你是哪个模型")
print(response.content)
```

# 2 模型调用

## 2.1 使用invoke调用

### 2.1.1 传入单条消息

```python
response = model.invoke("写一首关于秋天的五言诗")
print(response.content)
```

### 2.1.2 手动构造消息列表

```python
conversation = [
    {"role": "system", "content": "你是一个将英语翻译成法语的有用助手。"},
    {"role": "user", "content": "翻译：我喜欢编程。"},
    {"role": "assistant", "content": "J'adore la programmation."},
    {"role": "user", "content": "翻译：我喜欢构建应用程序。"}
]

response = model.invoke(conversation)
print(response.content)
```

### 2.1.3 使用消息类型构建消息列表

```python
from langchain.messages import HumanMessage, AIMessage, SystemMessage

conversation = [
    SystemMessage("你是一个将英语翻译成法语的有用助手。"),
    HumanMessage("翻译：我喜欢编程。"),
    AIMessage("J'adore la programmation。"),
    HumanMessage("翻译：我喜欢构建应用程序。")
]

response = model.invoke(conversation)
print(response.content)
```

## 2.2 使用stream调用

```python
full = None  # None | AIMessageChunk
for chunk in model.stream("天空是什么颜色？"):
    full = chunk if full is None else full + chunk
    print(full.text)

print(full.content_blocks)
```

### 2.2.1 流式传输事件

```python
async for event in model.astream_events("你好"):

    if event["event"] == "on_chat_model_start":
        print(f"输入：{event['data']['input']}")

    elif event["event"] == "on_chat_model_stream":
        print(f"令牌：{event['data']['chunk'].text}")

    elif event["event"] == "on_chat_model_end":
        print(f"完整消息：{event['data']['output'].text}")

    else:
        pass
```

## 2.3 使用batch调用

### 2.3.1 非流式返回

```python
responses = model.batch([
    "为什么鹦鹉有五颜六色的羽毛？",
    "飞机是如何飞行的？",
    "什么是量子计算？"
])
for response in responses:
    print(response)
```

### 2.3.2 流式返回

```python
for response in model.batch_as_completed([
    "为什么鹦鹉有五颜六色的羽毛？",
    "飞机是如何飞行的？",
    "什么是量子计算？"
]):
    print(response)
```

# 3 工具调用

```python
from langchain.tools import tool

@tool
def get_weather(location: str) -> str:
    """获取某个位置的天气。"""
    return f"{location} 天气晴朗。"

model_with_tools = model.bind_tools([get_weather])  # [!code highlight]

response = model_with_tools.invoke("波士顿的天气怎么样？")
for tool_call in response.tool_calls:
    # 查看模型发出的工具调用
    print(f"工具：{tool_call['name']}")
    print(f"参数：{tool_call['args']}")
print(response.content)
```

## 3.1 工具执行循环

```python
# 将（可能多个）工具绑定到模型
model_with_tools = model.bind_tools([get_weather])

# 步骤 1：模型生成工具调用
messages = [{"role": "user", "content": "波士顿的天气怎么样？"}]
ai_msg = model_with_tools.invoke(messages)
print(ai_msg)
messages.append(ai_msg)

# 步骤 2：执行工具并收集结果
for tool_call in ai_msg.tool_calls:
    # 使用生成的参数执行工具
    tool_result = get_weather.invoke(tool_call)
    print(tool_result)
    messages.append(tool_result)

# 步骤 3：将结果传递回模型以获取最终响应
final_response = model_with_tools.invoke(messages)
print(final_response.text)
# "波士顿当前天气为 72°F，晴朗。"
```
